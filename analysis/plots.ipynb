```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# runtime usr sys idl wai hiq siq used buff cach free recv send read writ memheap memdirect
data = np.load('tpchq10.npz')
sorted(data.keys())
```

```python
def plot_usage(X, col):
    
    num_machines, t = X.shape[:2] 
    
    plt.figure(figsize=(13, num_machines))
    for i in range(num_machines):
        plt.subplot(num_machines/4, 4, i+1)
        plt.plot(X[i,:,col])
    plt.tight_layout()
    plt.show()
    
    plt.plot(np.mean(X[:,:,col], axis=0)/1024.**2)
    plt.show()    
```

```python
plot_usage(data['top012.run01'], 7)
plot_usage(data['top012.run02'], 7)
```

```python
X = data['top012.run01']
for i in range(X.shape[0]):
    plt.plot(X[i,:,7]/1024.**2)
plt.show()    

plt.plot(np.mean(X[:,:,7], axis=0)/1024.**2)
plt.show()

for i in range(X.shape[0]):
    plt.plot(X[i,:,0])
plt.show()

X = data['top012.run03']
for i in range(X.shape[0]):
    plt.plot(X[i,:,7]/1024.**2)
plt.show()

plt.plot(np.mean(X[:,:,7], axis=0)/1024.**2)
plt.show()

for i in range(X.shape[0]):
    plt.plot(X[i,:,0])
plt.show()
```

```python
y = data['runtime']/1000./60.
x = np.arange(1, 16) * 4
X = np.c_[np.ones_like(x), 1./x]
c, _, _, _ = np.linalg.lstsq(X, y)

x_plot = np.linspace(1, 15, 50)*4
X_plot = np.c_[np.ones_like(x_plot), 1./x_plot]
y_predict = np.dot(X_plot, c)

plt.plot(x_plot, y_predict, 'k--', linewidth=2, label='Fitted Runtime Curve')
plt.plot(x, y, 'x', mew=2, label='Runtime')
plt.xlabel('# Nodes')
plt.ylabel('Runtime [m]')
plt.xticks(x)
plt.xlim(x[0], x[-1])
plt.ylim(ymin=0)
plt.legend()

plt.show()
```
